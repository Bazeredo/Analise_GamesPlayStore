```python
g_test = games.copy()
```

```python
plt.figure(figsize = (10,10))

ax= sns.barplot(x= 'title', y= 'total ratings', data= g_test,
               order= g_test.sort_values('total ratings', ascending = False).head(10).title, palette= 'Reds_r')
plt.title('Total Ratings', fontsize = 20)
plt.xticks(rotation= 90)
for patch in ax.patches:
    height = patch.get_height()
    width= patch.get_width()
    left_coord= patch.get_x()
    percent= height/sum(g_test['total ratings'])*100
    
    ax.text(left_coord + width/2, height + 700000,
        '{:.1f}M\n({:.1f}%)'.format(height/100000, percent), ha= 'center')
```

```python
#  4 e 5 estrelas é "bom" e o resto seria "ruim"
g_test['good ratings']= g_test['4 star ratings'] + g_test['5 star ratings']
g_test['bad ratings']= g_test['1 star ratings'] + g_test['2 star ratings'] + g_test['3 star ratings']

#calculando o percentual
g_test['good ratings percent']= g_test['good ratings'] / (g_test['good ratings']  + g_test['bad ratings'])
g_test['bad ratings percent']= g_test['bad ratings'] / (g_test['good ratings'] + g_test['bad ratings'])

g_test.head(2)
```

```python
sns.set_style('whitegrid')

plt.figure(figsize= (13, 10))

ax= sns.countplot(y= 'category', data= g_test, order= g_test['category'].value_counts().index, palette= 'GnBu_r')
plt.title('Category', fontsize= 20)

for patch in ax.patches:
    heigh= patch.get_height()
    width= patch.get_width()
    left_coord= patch.get_x()
    bot_coord= patch.get_y()
    percent= width/len(g_test)*100
    
    ax.text(left_coord + width, bot_coord + 0.7, '{}\n({:.1f}%)'.format(width, percent))
```
